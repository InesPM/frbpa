```python
from frbpa.plotting import make_obs_phase_plot
from frbpa.utils import get_params
import pylab as plt
%matplotlib inline
```

```python
params = get_params(1, 1.2)
plt.rcParams.update(params)
```

```python
make_obs_phase_plot('r3all_data.json', 16.26, ref_mjd=58369.30, nbins=40, save=False, show=True)
```

```python
make_obs_phase_plot('r3all_data.json', 16.35, ref_mjd=58369.30, nbins=40, save=False, show=True)
```

```python
make_obs_phase_plot('r3all_data.json', 16.35+0.1, ref_mjd=58369.30, nbins=40, save=False, show=True)
```

#### So even within a period error of 0.1, the phase distribution of the bursts doesn't change much. The high frequency (VLA, Effelsberg) bursts still lie at an earlier phase. The width of the activity window changes for different period, as expected.
